```python
# Show count of booking statuses
# Group by booking status and count the number of bookings
booking_status_counts = df.groupby('booking_status').size()

# Plotting
booking_status_counts.plot(kind='bar', rot=0)
plt.xlabel('Booking Status')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings by Status')
plt.show()

# Print the number of bookings that were canceled and not canceled
print("Number of bookings that were Canceled:", booking_status_counts['Canceled'])
print("Number of bookings that were Not Canceled:", booking_status_counts['Not_Canceled'])
```

```python
# price per room over date

df['avg_price_per_room'] = pd.to_numeric(df['avg_price_per_room'], errors='coerce')
df['room_type_reserved'] = pd.to_numeric(df['room_type_reserved'], errors='coerce')

# Handle missing values (replace NaN with 0, for example)
df[['avg_price_per_room', 'room_type_reserved']] = df[['avg_price_per_room', 'room_type_reserved']].fillna(0)

# Calculate price per room
df['price_per_room'] = df['avg_price_per_room'] / df['room_type_reserved']

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['price_per_room'], marker='o', linestyle='-')
plt.title('Price per Room Over Date')
plt.xlabel('Date')
plt.ylabel('Price per Room')
plt.grid(True)
plt.show()
```

```python
# show weekday by booking status

# Group by weekday and booking status, count occurrences, and unstack

# Create a cross-tabulation of 'day_of_the_week' and 'booking_status'
weekday_booking_counts = pd.crosstab(df['day_of_week'], df['booking_status'])

# Plotting
weekday_booking_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Weekday-wise Booking Status Distribution')
plt.xlabel('Weekday')
plt.ylabel('Number of Bookings')
plt.legend(title='Booking Status')
plt.show()

```

```python
 # show weekend nights by booking status
    # Group by 'no_of_weekend_nights' and 'booking_status', count occurrences, and unstack
weekend_nights_booking_counts = df.groupby(['no_of_weekend_nights', 'booking_status']).size().unstack(fill_value=0)

# Plotting
weekend_nights_booking_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Weekend Nights-wise Booking Status Distribution')
plt.xlabel('Number of Weekend Nights')
plt.ylabel('Number of Bookings')
plt.legend(title='Booking Status')
plt.show()
```

```python
# show booking status by month, day of the week and year

# Extract month, day of the week, and year from the 'date' column
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.day_name()
df['year'] = df['date'].dt.year

# Group by 'month', 'day_of_week', 'year', and 'booking_status', count occurrences, and unstack
booking_status_counts = df.groupby(['month', 'day_of_week', 'year', 'booking_status']).size().unstack(fill_value=0)

# Plotting for Month
plt.figure(figsize=(15, 8))
sns.heatmap(booking_status_counts.groupby(['month', 'booking_status']).size().unstack(fill_value=0), annot=True, fmt="d", cmap="YlGnBu")
plt.title('Booking Status by Month')
plt.xlabel('Booking Status')
plt.ylabel('Month')
plt.show()

# Plotting for Day of the Week
plt.figure(figsize=(15, 8))
sns.heatmap(booking_status_counts.groupby(['day_of_week', 'booking_status']).size().unstack(fill_value=0), annot=True, fmt="d", cmap="YlGnBu")
plt.title('Booking Status by Day of the Week')
plt.xlabel('Booking Status')
plt.ylabel('Day of the Week')
plt.show()

# Plotting for Year
plt.figure(figsize=(15, 8))
sns.heatmap(booking_status_counts.groupby(['year', 'booking_status']).size().unstack(fill_value=0), annot=True, fmt="d", cmap="YlGnBu")
plt.title('Booking Status by Year')
plt.xlabel('Booking Status')
plt.ylabel('Year')
plt.show()
```

```python
# show number of adults compared to children
# Group by 'no_of_adults' and 'no_of_children', count occurrences, and unstack
adults_children_counts = df.groupby(['no_of_adults', 'no_of_children']).size().unstack(fill_value=0)

# Plotting
adults_children_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Number of Adults vs Children Distribution')
plt.xlabel('Number of Adults')
plt.ylabel('Number of Bookings')
plt.legend(title='Number of Children')
plt.show()
```
